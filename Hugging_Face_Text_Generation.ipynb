```python
# prompt: rename columns with proper naming such as Content,Collection, Year of Publication

import pandas as pd
import numpy as np
poems = pd.read_csv('/content/robert_frost_collection.csv', on_bad_lines='skip')
#poems.head(5)
#print(poems.columns)

poems = poems.rename(columns={
    poems.columns[0]: 'Content',
    poems.columns[1]: 'Collection',
    poems.columns[2]: 'Year of Publication'
})

poems.columns
```

```python
content = poems['Content'].dropna().tolist()
```

```python
lines = []
for poem in content :
  for line in poem.split('\n'):
      lines.append(line.strip())
```

```python
lines = [line for line in lines if len(line) >0]
lines[:5]
```

```python
gen = pipeline("text-generation")
```

```python
lines[0]
```

```python
gen(lines[0],max_length = 20)
```

```python
gen(lines[1], max_length = 30, num_return_sequences = 2)
```

```python
from os import replace
import textwrap
def wrap(x) :
  return textwrap.fill(x, replace_whitespace = False, fix_sentence_endings=True)
```

```python
out = gen(lines[0], max_length = 30)
print (wrap(out[0]['generated_text']))
```
